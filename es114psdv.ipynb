##    1)Create a list of the first 20 natural numbers.
    Compute the sum of all even numbers in the list.
    Compute the product of all odd numbers in the list.
    Calculate the mean, median, and standard deviation of the list.


```python
naturalNo = list(range(1, 21))
totalSum=sum(naturalNo)
# sum of even
evenSum = sum(i for i in naturalNo if i & 1 == 0)

# product of odd
product = 1
for i in naturalNo:
    if i &1:
        product *= i
# mean
mean = totalSum / len(naturalNo)
sortedNo=naturalNo
# median
#sortedNo = sorted(naturalNo)  
n = len(sortedNo)
if n % 2 == 1:  
    median = sortedNo[n // 2]
else:  
    median = (sortedNo[n // 2 - 1] + sortedNo[n // 2]) / 2

# standard deviation
sq_sum = sum((i - mean) ** 2 for i in naturalNo)
stdev = (sq_sum / len(naturalNo)) ** 0.5

        
print(f"Natural Numbers: {naturalNo}")
print(f"Sum of Even Numbers: {evenSum}")
print(f"Product of Odd Numbers: {product}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {stdev}")
```

##  3)  Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
    Plot the Fibonacci sequence using a line graph.


```python
import matplotlib.pyplot as plt

def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]

    fib_sequence = [0, 1]
    for i in range(2, n):
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        fib_sequence.append(next_fib)
    
    return fib_sequence


def plot_fibonacci(n):
    fib_sequence = fibonacci(n)
    
    plt.figure(figsize=(12, 6)) 
    plt.plot(fib_sequence, marker='o', linestyle='-', color='skyblue', markersize=8, linewidth=2)  # Enhanced line style
    
    
    plt.title(f'First {n} Fibonacci Numbers', fontsize=18)
    plt.xlabel('Index', fontsize=14)
    plt.ylabel('Fibonacci Number', fontsize=14)
    plt.grid(color='gray', linestyle='--', alpha=0.7)
    plt.xticks(range(n), fontsize=12)  
    plt.yticks(fontsize=12)

  
    plt.tight_layout()  
    plt.show()
n = 20 
plot_fibonacci(n)
```

## 9)Write Python code to merge two dictionaries. If a key is present in both, sum their values.

Example:

dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
Output: {'a': 10, 'b': 25, 'c': 15}

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

merged_dict = dict1.copy()

for key, value in dict2.items():
    if key in merged_dict:
        merged_dict[key] += value
    else:
        merged_dict[key] = value

print(merged_dict)
```

## 11) Write a program to remove duplicate elements from a list while maintaining the order of elements.
    Calculate the mean and median of the unique elements.


```python
from statistics import mean, median

def remove_duplicates_and_calculate_stats(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    
    if unique_list:  
        calculated_mean = mean(unique_list)
        calculated_median = median(unique_list)
    else:
        calculated_mean = None
        calculated_median = None
    
    return unique_list, calculated_mean, calculated_median


input_data = [5, 3, 8, 3, 5, 10, 8, 1]
unique_elements, calculated_mean, calculated_median = remove_duplicates_and_calculate_stats(input_data)

print("Unique Elements:", unique_elements)
print("Mean of Unique Elements:", calculated_mean)
print("Median of Unique Elements:", calculated_median)
```

## 13. Write a function to sort a list of dictionaries by a key.
       - Example:
         - data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
         - Sort by age  [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]

```python
def sort_dict_list(data, key, reverse=False):
    
    return sorted(data, key=lambda x: x[key], reverse=reverse)


data = [
    {'name': 'Anu', 'age': 25}, 
    {'name': 'Dia', 'age': 22},
    {'name': 'Dev', 'age': 30}
]

# Sort by age (ascending)
sorted_by_age = sort_dict_list(data, 'age')
print(sorted_by_age)

# Sort by age (descending)
sorted_by_age_desc = sort_dict_list(data, 'age', reverse=True)
print(sorted_by_age_desc)

# Sort by name
sorted_by_name = sort_dict_list(data, 'name')
print(sorted_by_name)
```

## 15)Write a Python function to find all anagrams of a word in a given list.

    Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
    Output: ["enlist", "inlets"]


```python

def anagramfinder(word, words):
    sorted_word = sorted(word)
    return [w for w in words if sorted(w) == sorted_word]

word = "listen"
words = ["enlist", "google", "inlets", "banana"]
anagrams = anagramfinder(word, words)
print(anagrams)  # Output: ["enlist", "inlets"]

```

## 16)Given two lists, write Python code to find their intersection, union, and symmetric difference.
    Visualize the results using Venn diagrams
    


```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import numpy as np

def set_operations_visualization(list1, list2, title='Set Operations'):
  
    # Convert lists to sets
    set1 = set(list1)
    set2 = set(list2)

    # Compute set operations
    intersection = set1 & set2
    union = set1 | set2
    symmetric_difference = set1 ^ set2

    # Print set operations
    print(f"Intersection: {intersection}")
    print(f"Union: {union}")
    print(f"Symmetric Difference: {symmetric_difference}")

    #
    
    #  Venn Diagram
    plt.subplot(121)
    venn2([set1, set2], set_labels=('Set 1', 'Set 2'))
    plt.title('Venn Diagram')

    
   
    return {
        'set1': set1,
        'set2': set2,
        'intersection': intersection,
        'union': union,
        'symmetric_difference': symmetric_difference
    }

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

result = set_operations_visualization(list1, list2, 'Advanced Set Operations')
print(result)
```

## 18) Write a program to:

    
        Read a file and count the number of words in it.
        Find the most frequent word.
        Visualize the word frequencies using a bar chart.


```python
import matplotlib.pyplot as plt
from collections import Counter
import re

def analyze_text_file(file_path):

    try:
        
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read().lower()
        
       
        words = re.findall(r'\b\w+\b', text)
        
    
        word_counts = Counter(words)
        
     
        top_words = word_counts.most_common(10)
        
        # Visualization
        plt.figure(figsize=(12, 6))
        
        # Bar chart of word frequencies
        plt.subplot(121)
        words, frequencies = zip(*top_words)
        plt.bar(words, frequencies, color='skyblue')
        plt.title('Top 10 Word Frequencies')
        plt.xlabel('Words')
        plt.ylabel('Frequency')
        plt.xticks(rotation=45)
        
        
        
        plt.tight_layout()
        plt.show()
        
        # Return analysis results
        return {
            'total_words': len(words),
            'unique_words': len(set(words)),
            'most_frequent_word': word_counts.most_common(1)[0],
            'word_frequencies': dict(top_words)
        }
    
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None
    except Exception as e:
        print(f"Error processing file: {e}")
        return None

# Example usage
file_path = 'numbers.txt'
result = analyze_text_file(file_path)
print(result)
```

## 19)Write a Python program to:


    Generate 100 random integers between 1 and 50.
    Save them in a text file, one per line.
    Read the file and compute the mean of the numbers.
    Visualize the distribution of the numbers using a histogram.


```python
import random
import numpy as np
import matplotlib.pyplot as plt

def random_number_analysis(count=100, min_val=1, max_val=50):
    
    
    random_numbers = [random.randint(min_val, max_val) for _ in range(count)]
    
   
    output_file = 'random_numbers.txt'
    with open(output_file, 'w') as file:
        file.write('\n'.join(map(str, random_numbers)))
    
    
    with open(output_file, 'r') as file:
        file_numbers = [int(line.strip()) for line in file]
    
    
    mean_value = np.mean(file_numbers)
    median_value = np.median(file_numbers)
    std_dev = np.std(file_numbers)
    
    
    plt.figure(figsize=(12, 5))
    
    
    plt.subplot(121)
    plt.hist(file_numbers, bins='auto', edgecolor='black', alpha=0.7)
    plt.title('Random Number Distribution')
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    
    
    plt.tight_layout()
    plt.show()
    
    
    return {
        'random_numbers': file_numbers,
        'mean': mean_value,
        'median': median_value,
        'standard_deviation': std_dev,
        'min': min(file_numbers),
        'max': max(file_numbers)
    }


result = random_number_analysis()
print(result)
```

## 21)Write Python code to calculate:
    
    Mean, median, mode, and standard deviation of a list of numbers.
    Visualize the data using a box plot.


```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
from scipy import stats

def advanced_statistical_analysis(numbers):
    
   
    mean_value = np.mean(numbers)
    median_value = np.median(numbers)
    
    try:
        mode_value = statistics.mode(numbers)
    except statistics.StatisticsError:
        mode_value = "No unique mode"
    
    std_dev = np.std(numbers)
    variance = np.var(numbers)
    
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    iqr = q3 - q1
    
    # Visualization
    plt.figure(figsize=(12, 5))
    
    # Box Plot
    plt.subplot(121)
    plt.boxplot(numbers, vert=True)
    plt.title('Data Distribution Boxplot')
    plt.ylabel('Values')
    
    
    
    plt.tight_layout()
    plt.show()
    
    # Outlier Detection
    z_scores = np.abs(stats.zscore(numbers))
    outliers = [num for num, z in zip(numbers, z_scores) if z > 3]
    
    return {
        'mean': mean_value,
        'median': median_value,
        'mode': mode_value,
        'standard_deviation': std_dev,
        'variance': variance,
        'quartile_1': q1,
        'quartile_3': q3,
        'interquartile_range': iqr,
        'outliers': outliers
    }

# Example Usage
sample_data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 22, 18, 200]
result = advanced_statistical_analysis(sample_data)
print(result)
```

## 22)Write a Python function to normalize a list of numbers to a range of 0 to 1.Visualize the normalized data using a line plot.


```python
import numpy as np
import matplotlib.pyplot as plt


def normalize_data(numbers, method='minmax'):
  
  
    if method == 'minmax':
        normalized = (numbers - np.min(numbers)) / (np.max(numbers) - np.min(numbers))
    
 
    else:
        raise ValueError("Invalid normalization method")
    
    # Visualization
    plt.figure(figsize=(12, 5))
    
    # Original vs Normalized Line Plot
    plt.subplot(121)
    plt.plot(numbers, label='Original', marker='o')
    plt.plot(normalized, label='Normalized', marker='x')
    plt.title(f'{method.capitalize()} Normalization')
    plt.legend()
    
    
    
    plt.tight_layout()
    plt.show()
    
    return {
        'original_data': numbers,
        'normalized_data': normalized,
        'min_value': np.min(normalized),
        'max_value': np.max(normalized),
        'method': method
    }

# Example Usage
sample_data = [27, 10, 45, 45, 19, 34, 40, 40, 5]

# Different Normalization Methods
print("Min-Max Normalization:")
minmax_result = normalize_data(sample_data, method='minmax')
print(minmax_result)

```

## 25)Write a Python program that takes a year as input and checks if it is a leap yea
